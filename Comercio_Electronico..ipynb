```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
```

Extraído de: https://www.inegi.org.mx/investigacion/nei/#tabulados

```python
Comercio = "C:\\Users\\sc las alamedas\\Repoaitorios\\Listings_Data\\Operaciones_Comercio_Electronico.xlsx"
```

```python
df = pd.read_excel(Comercio)
df.head(10)
```

```python
# Encuentra el índice de la fila que contiene 'Fecha' en la columna 'Col1'
index_fecha = df[df['Banco de México'] == 'Fecha'].index[0]
# Usa la fila que contiene 'Fecha' como nuevo encabezado, sin el índice
new_header = df.iloc[index_fecha]
df = df[index_fecha + 1:]  # Filtra las filas del DataFrame a partir del índice de 'Fecha' + 1
df.columns = new_header  # Asigna la fila 'Fecha' como encabezado
df = df.reset_index(drop=True)  # Restablece el índice
```

```python
# Mostrar el DataFrame filtrado
df.head(10)
```

```python
df.info()
```

```python
# Convertir la primera columna a tipo fecha
df.iloc[:, 0] = pd.to_datetime(df.iloc[:, 0])

# Convertir todas las demás columnas a tipo flotante
for col in df.columns[1:]:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.info()
```

```python
# Registrar los conversores explícitamente
register_matplotlib_converters()

# Crear la gráfica de área
plt.figure(figsize=(12, 6))
plt.fill_between(df['Fecha'], df['SF273475'], color='skyblue', alpha=0.4)
plt.plot(df['Fecha'], df['SF273475'], color='blue', alpha=0.6)
plt.title('Área de SF273475 a lo largo del tiempo')
plt.xlabel('Fecha')
plt.ylabel('SF273475 (millones)')

# Configurar el formato del eje Y para mostrar los números en millones y sin notación científica
formatter = plt.FuncFormatter(lambda x, _: '{:1.1f}M'.format(x / 1e6))
plt.gca().yaxis.set_major_formatter(formatter)

plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
